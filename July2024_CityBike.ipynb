Creating united colums with data for July 2024

```python
#Lets' unite all July records of 2024 
Jul2024_df = pd.concat([
    jul2024_page1_df, 
    jul2024_page2_df,
    jul2024_page3_df,
    jul2024_page4_df,
    jul2024_page5_df
    ], 
    ignore_index=True
    )

Jul2024_df.info()
```

```python
#our colums starttime and stoptime are 'object' 
# we need it to be in datetime format
Jul2024_df_edt=Jul2024_df.copy()
```

```python
# started_at and ended_at should be datetime format

Jul2024_df_edt["started_at"] = pd.to_datetime(Jul2024_df_edt["started_at"]).dt.floor('S')
Jul2024_df_edt["ended_at"] = pd.to_datetime(Jul2024_df_edt["ended_at"]).dt.floor('S')
Jul2024_df_edt.info()
```

```python
#Let's creat trip duration for July 2024 df
# Create the trip duration in minutes:seconds format
Jul2024_df_edt['trip_duration'] = (Jul2024_df_edt['ended_at'] - Jul2024_df_edt['started_at'])

# Calculate the trip duration in minutes (numeric format)
Jul2024_df_edt['trip_duration_minutes'] = (
    (Jul2024_df_edt['ended_at'] - Jul2024_df_edt['started_at']).dt.total_seconds() / 60
).round(2)

# Check the result
Jul2024_df_edt[['started_at', 'ended_at', 'trip_duration', 'trip_duration_minutes']].head(10)
```

```python
#Let's unify columns' names
Jul2024_df_edt.rename(columns={
    'started_at': 'start_time',
    'ended_at': 'end_time',
    'member_casual': 'membership'
}, inplace=True)

# Check the column names
Jul2024_df_edt.columns
```

```python
#checking Jul2024_df_edt
Jul2024_df_edt.info()
```

```python
#Let's check membership cells
Jul2024_df_edt["membership"].value_counts()
```

```python
#renaming membership cells to "member" and "non-member"
Jul2024_df_edt['membership'] = Jul2024_df_edt['membership'].replace(
    {'casual': 'non-member'
     })

#checking the succsess 
Jul2024_df_edt["membership"].value_counts()
```
